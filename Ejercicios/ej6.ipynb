### Preguntas:

- ¿Por qué la norma L1 representa la suma absoluta de las calificaciones? <br>
Porque la norma L1 en definición es la suma obsoluta de todos los componentes de un array, el cual en nuestro caso son las columnas de nuestra matriz. <br>

- ¿Por qué la norma L2 nos da una idea de la magnitud total de los valores, penalizando más las notas altas?
- ¿Cómo afecta `keepdims=True` en las operaciones realizadas? <br>
Mantiene las dimensiones de los vectores utilizados en las operaciones en las que se utiliza. Por ejemplo, si yo aplico la norma L2 en cada fila de una matriz 3x3 sin aplicar 'keepdims' me devolverá los resultados en una matriz de 1x1, pero si por el contrario aplico dicho parámetro el resultado mantendrá las dimensiones de la matriz original (3x3)
